# Nested For Loops

```python
flavors = ['Chocolate', 'Vanilla', 'Straeberry']
toppings = ['Oreos', 'Berry', 'Beans']
```

```python
for one in flavors:
    for two in toppings:
        print(one, 'topped with', two)
```
