# If- Elif - Else Statements

```python
if 25 > 20:
    print('Berhasil!')
```

```python
if 25 < 20:
    print('Berhasil!')
else:
    print('tidak berhasil')
```

```python
if 25 < 20:
    print('Benarrrr!')
elif 25 < 20:
    print('ini juga 1 benar!')
elif 25 < 21:
    print('ini juga 2 benar!')
elif 25 < 40:
    print('ini juga 3 benar!')
elif 25 < 50:
    print('ini juga 4 benar!')
else:
    print('tidak benar')
```

# Latihan penerapan

```python
#input nilai siswa
nilai = float(input('Masukan Nilai'))

#menentukan peringkat siswa berdasarkan nilai

if nilai >= 90:
    peringkat = 1
elif nilai >= 80:
    peringkat = 2
elif nilai >=70:
    peringkat = 3
elif nilai >= 60:
    peringkat = 4
else :
    peringkat = 5

print('nilai siswa', nilai)
print ('peringkat siswa', peringkat)
```
