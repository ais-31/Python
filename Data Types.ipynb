```python
daily_pints_log = {1,2,31,2,3,4,1,2,5,6,3,2}

print(daily_pints_log)
```

```python
wife_daily_pints_log = {1,3,5,7,3,24,5,7,3,2,0}
```

```python
#Gabunga
print (daily_pints_log | wife_daily_pints_log)
```

```python
#Irisan
print (daily_pints_log & wife_daily_pints_log)
```

```python
#Selisih
print (daily_pints_log - wife_daily_pints_log)
```

```python
#hanya salah satu
print (daily_pints_log ^ wife_daily_pints_log)
```
